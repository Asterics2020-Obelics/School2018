```python
import pandas as pd
```

```python
df = pd.read_html("https://indico.in2p3.fr/event/16864/page/1800-list-of-participants",
            match="ASTRON", index_col=0, header=0, encoding='utf-8')[0]
df
```

```python
import geocoder
```

```python
def calculate_geocode(row):
    return geocoder.komoot(f"{row['Country']}, {row['City']}").latlng
```

```python
df['LatLong'] = df.apply(calculate_geocode, axis=1);
```

```python
import folium
```

```python
participants_map = folium.Map(location=[48.91494834, 6.86876941], zoom_start=4)
```

```python
def add_participant(row):
    folium.Marker(row['LatLong']).add_to(participants_map)
```

```python
df.apply(add_participant, axis=1);
```
